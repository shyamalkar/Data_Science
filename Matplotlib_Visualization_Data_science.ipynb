***`Line Plot`***

Line Plot generally usage for :-   
               time series data

               trend analysis

               stock price change

               temperature change

`E.g., `

```python
from matplotlib import lines
import matplotlib.pyplot as plt

months = ['jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun']
recovery_rate = [65, 70, 60, 78, 80, 85]
plt.plot(months, recovery_rate, marker='o', linestyle='-', color='b', label='recovery Rate')

plt.xlabel('Months')
plt.ylabel('Recovery Percentage')
plt.title("Patient Recovery Trends")
plt.legend()
plt.plot()
plt.show()
```

```python
medicines = ['Peracetamol', 'Antibiotics', 'Painkillers', 'Insulin', 'Antacids']
demand = [500, 700, 450, 300, 650]
plt.bar(medicines, demand, color=['red', 'blue', 'green', 'purple','orange'])

plt.xlabel('X-Label')
plt.ylabel('y-label')
plt.title("Medicine demand analysis")
plt.legend()
plt.plot()
plt.show()
```

***`Pie chart`***

    A pie chart is a circular graph that divides data into slices to show the proportion or percentage of each category.

    Usage in datascience:-

    percentage comparison

    category distribution

    population share

    market share

```python
import matplotlib.pyplot as plt

labels = ['Male','Female']
sizes = [60,40]

plt.pie(sizes, labels=labels, autopct='%1.1f%%')
plt.show()
```

***`Scatter Plot`***

   `A scatter plot is a graph that uses dots to represent the relationship between two variables.
   `

   Usage in Data science:-

    correlation check

    relationship between variables

    outlier detect

    regression analysis

```python
import matplotlib.pyplot as plt

x = [1,2,3,4,5]
y = [2,4,5,4,5]

plt.scatter(x,y)
plt.show()
```

# ***`Problem statement:`***

Problem statement :    
   Scatter Plot: Doctor patient interaction time

   A hospital Examines how time spent by doctors affects giagnosis accuracy

   We Visualize this relationship with scatter plot

   

```python
time_spent = [10, 15, 20, 30 ,40, 50, 60]
diagnosis_accuracy = [50, 60, 70 ,80, 90, 20, 30]

plt.scatter(time_spent, diagnosis_accuracy, color='red')

plt.xlabel("Time spend by doctor in minutes")
plt.ylabel("Diagnosis Accuracy (%)")

plt.title("Doctor-Patient Interecton time vs Diagnosis accuracy")
plt.show()
```

 ***`Histogram`***

    A histogram is a graph that shows the frequency distribution of numerical data by dividing it into intervals (bins).

    Data science usage:-

    data distribution

    frequency

    skewness detect

    outliers identify

```python
import matplotlib.pyplot as plt

data = [10, 12, 15, 20, 22]
bata = [23, 25, 30, 35, 40]

plt.hist([data, bata], bins=10)
plt.show()
```

***`Word Cloud `***

```python
from wordcloud import WordCloud

symptoms_text = "fever cough cold headache nausea fatigue pain weakness dizziness allergy infection sore throat congestion"
wordcloud = WordCloud(width=800, height=400, background_color='white').generate(symptoms_text)

plt.figure(figsize=(10, 5))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.title("Common symptoms Analysis")
plt.show()
```

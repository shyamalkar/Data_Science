# ***`Consider an insurence company analyzing claims made by policyholders`***

```python
import numpy as np

#Monthly claim amounts (in INR) for different policy holders

claims = np.array([
    [12000, 15000, 18000], #Policyholder 1 # axis = 1
    [10000, 13000, 17000], # Policyholder 2 #axis = 2
    [9000, 11000, 16000] # Policyholder 3 #axis = 3
])
print("insurance Claim Data (INR):")
print(claims)
```

```python
#NumPy allows performing  operatinns on entire arrays efficiently
#Example - Total claim Amount Per Policyholder

#Calculate total Claim per policy holder

total_claims = np.sum(claims, axis=1) #1 for row wise, 0 for column wise

print("Total Claims per Policyholder:", total_claims)
```

# ***Case Study: Policyholder data Transformation in insurance industry using numpy***

``Overview``
     An insurance company stores policyholder data, including policy number, premium amount, insured amount and claim amount.

     Due to a data migration issue, the "Premium amount" And "Claim Amount" columns have been swapped .Your track is to crrest this misaligment using Numpy.


``Learning Objectives:``
     
     Understand how to manipulate structure data using NumPy
     perform column swaping in a Numpy array.
     Gain Insights into real-world insurence data handling .

  

```python
import numpy as np
import pandas as pd

#Step 1 : Create a sample sataset representing insurance policyholder records

#Columns plicy Number, premium amount (Incorrect), Insured. Amount, Claiim Amount(Incorrect)

policy_data = np.array([
    [1001, 12000, 500000, 5000],
    [1002, 15000, 750000, 7000],
    [1003, 18000, 1000000, 10000],
    [1004, 20000, 1200000, 12000],
    [1005, 25000, 1500000, 15000]
])

col = ["Policy Number", "premium Amount", "Insured Amount", "Calim Amount"]

#Print Original Dataset
print("Original Policy Data:")
print()

def_before = pd.DataFrame(policy_data, columns=col)
print(def_before)

#Policy Number is my DataFrame , and now we gonna convert swapped premium amount column go to the last and last column come 2nd position.

policy_data[:, [1,3]] = policy_data[:, [3, 1]]

#We use Numpy slicing ('[:, [1, 3]] = [:, [3, 1]]') to swap the misplaced columns efficiently

print("\n Corrected Policyholder Data (After Swapping premium and claim amount columns):")
print()

df_after = pd.DataFrame(policy_data, columns=col)
print(df_after)

#Now You can see premium amount column value are change and claim amoutn value number are store in premium amount.

```

```python
## Real World Impact :
## - Insurence companies uselarge for premium calculations and claim processing.
## - Incorrect column placement can lead to incorrect financial reporting .
## This technique ensures **dataaccuracy** before feeding it into analytics or decision-making system.
```

```python
#Additional Task: Calculate Total Claim Amount Processed
total_claim_amount = np.sum(policy_data[:, 3])
print(f"total claim amount:-", total_claim_amount)
```

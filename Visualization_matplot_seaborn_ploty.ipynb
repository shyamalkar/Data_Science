# ***Tourism Trends Over Time***

```python
import pandas as pd

import plotly.express as px

df = pd.read_csv('tourism_data.csv')
df
```

```python
import pandas as pd
import plotly.express as px

df_year = df.groupby("Year")["Revenue_Billion_USD"].mean().reset_index()

fig = px.line(df_year, x='Year', y='Revenue_Billion_USD', title='Tourist Revenue by year')
fig.show()
```

***`Bar Plot`***
    
    Comcept:-

          Tourist arival by Country

          

```python
fig = px.bar(df, x='Country', y='Visitors_Millions', title='Tourist arrivals By Country')
fig.show()
```

# ***Histogram***
 Concept
        
        A histogram shows the frequency distribution of numerical data, making it useful for analyzing hotel pricing .


        Distribution of hotel prices

```python
fig = px.histogram(df, x='Average_Rating', title='Average Rating by hotel')
fig.show()
```

# ***Violin Plot***
     
              Concept:- A violin plot combines a box plot and KDE (Kernal Density. Estimation) to show data distribution with density extimation.

              

# ***Year by visitor in Millions***

```python
fig = px.violin(df, x='Year', y='Visitors_Millions', box=True, title='Year by millions visitor')
fig.show()
```

# ***Heatmap***
      
     Concept:-
            
            A heat map is useful for finding patterns and correlations.
            



# ***Correlation Between Travel Features***

```python
import plotly.figure_factory as ff
import numpy as np
```

```python
import plotly.figure_factory as ff #Ploty is a  interactive visuaization library.uses figure_factory able to create a special charts.

#Compute correlation matrix (numerical columns only)
correlation_matrix = df.select_dtypes(include=['number']).corr() #Both work done here, numeric column select . means chose only numeric columns.
#correlation_matrix = .corr() this is calculate numeric column - column relationship . and correlation result between -1 - +1 and 0 indicate no linear and 1 indicates a perfect .
#Create annotated heatmap with smaller font

fig = ff.create_annotated_heatmap( #Create a annotated_heatmap
    z=correlation_matrix.values, #Heatmap value , means heatmap color value , 1.00 , -0.45 ,0.32
                                   #x=correlation_matrix.columns.tolist() , column names and heat map show in axis. heatmap color theme. showscale=True , colorlegend showing on side .
    x=correlation_matrix.columns.tolist(),
    colorscale='Viridis',
    showscale=True,
    annotation_text=[[f"{val:.2f}" for val in row] for row in correlation_matrix.values], #Heat map write an number inside cell . 0.34, -0.51, 1.00 . :.2f means 2 decimal place .
    font_colors=['White'] #font_colors=['White'] means inside heatmap number are present in white.
)

#Update layout for spacing and readable text
import matplotlib.pyplot as plt

plt.figure(figsize=(4,4))

fig.update_layout(  #Modify the graph layout .
    width=800, #Graph size
    height=800, # Graph size
    margin=dict(t=50, l=100), # space around graph
    font=dict(size=10) # text size
)

fig.show()
```

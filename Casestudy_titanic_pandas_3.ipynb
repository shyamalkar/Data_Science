# ***`Column Selection and Filtering`***
``Select columns of interest``

```python
#Column selection means take a specific column name

df["Age"] #It show only column for multi columns df[["Age", "Sex"]]


```

```python
#Filtering means select the data use condition .

df[df["Age"] > 30] # Means Upper 30 age
```

***Another Example:-***

```python
df[df["Sex"] == "female"] # All female passenger , Only chose Female passenger .
```

***`Multiple Condition Filtering`***

```python
#AND condition
df[(df["Age"] > 30) & (df["Sex"] == "female")] # Age > 30 AND female

#OR condition
df[(df["Age"] > 30) | (df["Sex"] == "female")] # Age > 30 OR female, in short female name upper 30 age .

#Column + Filter togather
df[df["Age"] > 30][["Name","Age"]] # Age > 30 , Only Name and column
```

# ***`Fill missing Base_encounter_cost values with the median`***

```python
df['Age'].fillna(df['Age'].median(), inplace=True) #df['Age'] means select the age column from dataset. here NaN means missing value, Calculate the median from Age column means middle value .fillna means replace missing value and fill median.
#Why median ?
#Because median is best and better than mean , If any outlier are present in dataset then mean creat so many problem.
```

# ***`What is Aggregation ?`***

     ``Aggregation means: So many data onvert into a summary value``

     e.g.,
          
          | Dept | Salary |
          | ---- | ------ |
          | IT   | 50000  |
          | IT   | 60000  |
          | HR   | 40000  |
          | HR   | 45000  |
      
      Here if we calculate average salary through Dept this is aggregation .

      What is GroupBy ?
          
          Group by means , Data group by columns .

          e.g,

          IT group → 50000, 60000
          HR group → 40000, 45000

          Mean means average

```python
df.groupby("Sex")["Age"].mean() #Average between Sex and Age
```

# ***Matrix Plot :-***
             
             Definition:-
                   
                   A matrix plot is a visualization that displays the relationship
                   between multiple variables in a dataset using a matrix format.



```python
corr = df.corr()

sns.heatmap(corr, annot=True, cmap="coolwarm")

plt.title("Correlation Matrix")
plt.show()
```

***Weather Analysis:- Heatmap of temparature variations over time.***

``‼️ Alert``
     
     Correlation values always between -1 or +1

     e.g.,

     A and B

     means:-

     1 means Strong Positive correlations

     if A increase then B increase .

     S.No and B 0.99 means almost perfect positive relation

     B-E moderate positive correlation


     Color means :-   
                
                Understand correlation with color .


                Color	               Meaning
                 🔴 Red	          Strong positive
                 🔵 Blue	       Negative correlation
                 Light color	   Weak relation


```python
plt.figure(figsize=(8,6))

#Compute correlation for numeric columns only

numeric_df = df.select_dtypes(include=['number']) #Select only numeric columns

sns.heatmap(numeric_df.corr(), annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)

plt.title("Correlations Matrix of Features")
plt.show()
```

***Box-Plot***
  ``Defination:-``
             
             A box plot is a graphical representation that shows the distribution of numerical data using quartiles, median and outliers.

``What does it show Boxplot ?``
              
             

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

data = {
    "Student": ["A","B","C","D","E","F","G","H"],
    "Marks": [45, 50, 55, 60, 65, 70, 75, 90]
}

df = pd.DataFrame(data)

print(df)
```

```python
plt.figure(figsize=(6,4))

sns.boxplot(data=df, y="Marks")

plt.title("Student Marks Distribution")
plt.show()
```

***Violin Plot – Definition***
    
    A violin plot is a visualization that shows the distribution of data
    and its probability density along with summary statistics.

``what does it show violine plot ``

         Part              Meaning

         Width             where data increaes
         Middle line.          median
         Shape	            distribution
         Spread            how many data spread
         

```python
data = {
    "Student": ["A","B","C","D","E","F","G","H"],
    "Marks": [45, 50, 55, 60, 65, 70, 75, 90]
}

df = pd.DataFrame(data)

sns.violinplot(y=df["Marks"])

plt.title("Marks Distribution")
plt.show()

#Where is more data ?
#How read Violin Plot ?
# Where is the data more ?
# Where is the data less?
# where is Median ?
# How many data is spread ?


#-:Understanding:-

                #Where shape Fat  → More
                #Where shape thin → data min

                #My shape start fat 55 and end aroung 70 . Means more marks in this range

                #Middle White dot - white dot means median
                #my graph median almost - 62 - 63

                #Black box = Q1 to Q3 (25% - 75%)
                #Means most marks are situated in this range .
                #Top and bottom line
                #Top = Maximum value
                #Bottom = Minimum Value
                #My data = Min = 45, Max = 90 small black line in middle .


                #What is the meaning of Quartile ?
                #Data divide into 4 parts.

                #Quartile                    Meaning

                #Q1                           25% data

                #Q2                           50% data (Median)

                #Q3                           75% Data
#Diagram
    #Min      Q1        Q2        Q3       Max
    #|--------|---------|---------|--------|
    #45       52.5      62.5      72.5      90


#Visual understand:- 45  50 | 55  60 | 65  70 | 75  90
#                          Q1       Q2       Q3

#Here:-
#Q1 = 25%
#Q2 = median
#Q3 = 75% data
```

# ***`EDA(Exploratory Data Analysis)`***

      Means:-
           
           Explore the data and understand what is the real meaning in data .

           in short:-
                   
                   Process of understand the data . when Data Scientist and analyst gets data first time then they do EDA first for understand better .

                   What is the real meaning of this data ?

                   Have a missing value or not.

                   Outlier or not ?

                   Which Feature is importent ?
                   
                   What is the relation between feature .


``Why invent EDA ? ``
                 
                 Before it's very hard to understand the data and one company collect 10 lakhs customer data. But they don't know

                 Data clean or not ?
                 Data accurate or not ?
                 Column means
                 what is the data distribution .

                 If you predict without clean the data, model give bad response, garbage result -> garbage data

                 So yeahh ! this is the reson why EDA was invent .


``Why use EDA ? ``
                 
                 Data structure understand
                 Missing value detect
                 Outlier Detect #E.g., 23, 45, 34, 23, 24, 56, 400 . So 400 is a Outlier .
                 Feature Relationunderstand
                 Understand the data distribution .
                 Deature selection

``Where use EDA ?``
                 
                 Machine learning
                 Data science
                 Business analytics
                 research

``Main parts in EDA  ``
                 
                 Univariate Analysis → 1 variable

                 Bivariate Analysis → 2 variable

                 Multivariate Analysis → 3+ variable

# ***`Used Car Price Analysis using Matplotlib and Seaborn`***

``Dataset: Used car Listing from CarDekho``

                ``Objective:-`` Solve business problems visually - price trends, feature effects, resale value patterns.


``Step 1: Imports and Configuration``



```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Set theme
sns.set(style='whitegrid')
plt.rcParams['figure.figsize'] = (12, 8)

```

```python
#Step 2 : Load the dataset .
import kagglehub

# Download latest version
path = kagglehub.dataset_download("manishkr1754/cardekho-used-car-data")

print("Path to dataset files:", path)
```

```python
import os
os.listdir(path)
```

```python
import pandas as pd

df = pd.read_csv(path + "/cardekho_dataset.csv")
df
```

# ***Step 3:- Fuel type Distribution***

```python
  #Fuel type availability
  #sns.countplot(x='fuel', data=df, palette='Set2')
  sns.countplot(x='fuel_type', hue='fuel_type', data=df, palette='Set2', legend=True) # This is actually a ploting line . if you use True then is show legend box, if false then it is not able to show legend list .
  #sns.countplot(), usage this to show categorical data count. sns.countplot(0) means how many times come every each an category .
  #x='fuel_type' means fuel type column showing in x axis . in my dataset fuel_type - Petrol, Diesel, CNG, LPG, Electric.
  #data=df, dataset
  #hue='fuel_type' means changing color according to fuel_type
  # palette='Set2'means color theme .
  #legend=False, never show legend box .

  plt.title('Fuel Type Distribution')
  plt.show()
```

# ***`Step 4: Selling price Distribution`***

```python
df.head()
```

```python
# * price range analysis
plt.figure(figsize=(20, 20))
sns.histplot(df['selling_price'], kde=True, bins=100, color='skyblue') #color = 'skyblue' means color blue
plt.title('Selling Price Distribution')
plt.xlabel('Price(INR)')
plt.show()

#Visualization Meaning
#What is this graph ?this is  Histogram + KDE distribution plot, This is show us How to distribute car selling price dataset ?
#code explanation
#kde=True,  smooth distribution curve , price range deivide by 30 step .
# bins=30 means , what does mean actually ?
#   If my price data :-
#100
#200
#300
#400
#500
#600
#700
#800
#900
#Minimum price
#100
#Maximum Price
#900
#Total Range:
#900-100 = 800
#IF bins give 4 then price range divide by 8 step . every bar is a single bin .
```

# ***Step 5: Price by Fuel Type (BoxPlot)***

```python
#Fuel vs price
sns.boxplot(x='fuel_type', hue='fuel_type', y='seller_type', data=df,  palette='Set1')

plt.title('Price Distribution by Fuel Type')
plt.show()
```

## *Price by Transmission*

```python
#Transmission impact
sns.boxplot(x='km_driven', y='vehicle_age', hue='transmission_type', data=df, palette='coolwarm')
plt.title('Selling Price by Transmission Type')
plt.show()
```

# ***`Step-7:- Price by Ownership History`***

```python
# * Owner impact
plt.figure(figsize=(6,6))
sns.violinplot(x='car_name', hue='car_name', y='selling_price', data=df, palette='Set3')
plt.title('Selling Price vs Owner History')
plt.show()
```

# ***`Correlation Heatmap`***

```python
correlation = df[['vehicle_age','km_driven', 'selling_price']].corr()
sns.heatmap(correlation, annot=True, cmap='viridis')
plt.title('Feature correlation Heatmap')
plt.show()
```

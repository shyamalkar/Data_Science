# **`Min Max Scaler`**
     
     Min-Max Scaler

        What is min max scaler ?
           
           min max scaler usage for data represent under 0-1

           e.g.,
              
              Salary - 20000, 45000, 60000, 120000

              After Scaling ->
                            
                            Salary ->
                                   
                                    0.00, 0.25, 0.40, 1.00 .

                                    Lowest value -> 0 , highest -> 1


           Formula :- x' = x - x_min/ x_max - x_min
                
                Means :-   Salary = [20000, 45000, 60000, 120000]

                Minimum value
                      
                      20000

                Max value 120000

            
            Step -1 (20000)

               (20000 - 20000) / (120000 - 20000)
               = 0 / 100000
               = 0

            Step -2 (45000)

            (45000 - 20000) / 100000
            = 25000 / 100000
            = 0.25

            Step -3

            (60000 - 20000) / 100000
            = 40000 / 100000
            = 0.40

            Step - 4

            (120000 - 20000) / 100000
            = 100000 / 100000
            = 1


````python
minmax_scaler = MinMaxScaler() # We create a Object in MinMaxScaler , Means now we can scale the data using minmax_scaler.

scaled_minmax = minmax_scaler.fit_transform(numeric_data) # minmax_scaler is a function and fit usage for see the data
print('\n Min-Max Scaled Data :\n', pd.DataFrame(scaled_minmax, columns=numeric_data.columns)) # scaled_minmax this is numpy array . this array convert into dataframe. columns=numeric_data.columns it is very importent because this is usage for original dataset copy and copy the column name .

print('\nReal dataset:\n',numeric_data)
````

One Line answer :--
                  Label Encoding converts categories into a single numeric column, while One-Hot Encoding creates separate binary columns for each category to avoid ordinal relationships.



Importent:-

Label Encoding is often used for the target variable (y) in classification problems, while One-Hot Encoding is commonly used for categorical feature variables (X) to avoid introducing ordinal relationships.


# ***`Normalization`***

       What is Normalization ?
             
             Purpose:-
                 
                 Scales each row to have a unit magnitude (length = 1) while preserving direction.


             Key Idea:-
                 Converts row vector into the pure direction from by dividing by their euclidean norm .

             When to use ⁉
                       
                       Text data (Tf-IDF word counts)

                       Clustering (K-means, cousine similarity)

                       Any algorithm sensitive to vector magnitudes but not theirlength (NLP, Recommender system)
              
                       

````python
from sklearn.preprocessing import Normalizer

#Example data

X = np.array([[3.0, 4.0], [1.0, 0.0], [0.0, 8.0]])
X
````

````python
X
````

````python
normalizer = Normalizer()
normalizer_data = normalizer.fit_transform(X)
print('\n Normalized Data: \n', normalizer_data)
````

***`How It Works`***

For each row , the normalizer . Computes the Euclidean norm(magnitude)

             norm = sqrt(x one square + x two square + ...)

             Divides each element by the norm:

             x one normalized = x one / norm, x square normalized = x 2 / norm

             calculates row Norm : for [3,4] - >
             norm = root(3 square + 4 square) = root(9 + 16) = 5



             Normalized

             DividesEach Element by Norm: [3/5, 4/5] = [0.6, 0.8].

             Second row :-  [1.0, 0.0] ->
             norm = root(1 square + 0 Square) = 1
             Normalized: [1/1, 0/1] = [1.0, 0.0]

             Thired row :-  [0.0, 8.0]:

             Norm = root ( 0 square + 8 square ) = 8

             Normalized: [0/8, 8/8] = [0.0, 1.0]

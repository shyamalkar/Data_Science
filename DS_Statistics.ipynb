 **``What is Simple Random Sampling ? What is Stratified Sampling ? What is Cluster Sampling ? ``**
            
            Simple Random Sampling .
                   
                   Simple Random Sampling is a method where use for Every number has an equal chance of being selected .

           Example
                 
                 Population = 100 students in a class.

                 i write all 100 names on paper, mix then , and randomly pick 10 names .

                 those 10 students - sample .

                 Every student had the same probability to be selected .

                 Real life use .

                    Random survey
                    Random data sampling in ML

        Stratified sampling :
                
                Defination :-   
                            
                            Population is Divided into groups(Strata) based on same characteristic.

                            Then we take samples from each group .

                            Example:-   School Population = 100 students .

                            60 Boys
                            40 girls

                            If you want to sample 20 students .

                            You select

                            12 boys
                            8 girls

                            This keeps the same proportion as the population.

                            This is stratified Sampling .

                            Why used ?

                            To make sure every group is represented

                Population is divided into clusters, usually based on location or natural groups.

                Then entire clusters are selected randomly .

                     A city has 50 school .

```python
from fractions import Fraction # This importsthe fruction class from python fractions module.
#Fraction use for Instead of geeting probabilities like:
#0.5 , decimal number transfer into number .

#Define the sample space and Outcomes

coin_outcomes = ['Heads', 'Tails'] # This defines all possible outcomes of a coin toss.
total_outcomes = len(coin_outcomes) # Total possible outcome is 2

#Calculate probabilities dynamically
prob_heads = Fraction(coin_outcomes.count('Heads'), total_outcomes) #First Python counts : 1r
prob_tails = Fraction(coin_outcomes.count('Tails'), total_outcomes)

print('Sample space:', coin_outcomes) # head and tail
print('P(heads) =', prob_heads) # Output: P(Heads) = 1/2
print('P(Tails) = ',prob_tails) # Output: P(Tails) = 1/2
```

# ***`Hypothesis Testing`***
        
        A method to make a decision using data.

        or:-

        Hypothesis testing is a statistical method used to determine if sample data provides enough evidence to reject a specific claim about a population parameter

        Generally :-   
         
         We check if a claim is true or not .

``Two types of Hypotheses are here``
          
          Null Hypothesis (H₀)
          Alternative Hypothesis (H₁ / Hₐ)

    
    What is Null Hypothesis ? (H0)
          
          A null Hypothesis (H0) is a fundamental statistical statement assuming there is no effect
          , no relationship, or no difference between measured variables or groups. It acts as the default position
          (status quo) to be tested, suggesting any observed differences are due to
          chance.
          Researchers aim to either reject or fail to reject


    What is Alternative Hypothesis ? (H0)
         
         An alternative hypothesis ( Ha or H1) is a statement in statistical testing
         that contradicts the null hypothesis (H0), proposing that there is a
         significant effect, difference,
         or relationship between variables. It represents the researcher's claim and is accepted if evidence allows the rejection of the null hypothesis.

      What is P-Value ?  
         
         A p-value is a number that tells you how likely it is that your data happened by total accident .
         
         p-value is not a separate method .

         it is a part of hypothesis testing .



      What is Z-Test ?
           
           Sample size is large ( n >= 30 ) or population std (σ) is known .

           forluma :- Z = (X̄ - μ) / (σ / √n)

           Meaning is :- Z = how many standard errors away your sample mean is .

           Example :-   
              
              Claim: Average salary = 5,00, 000

              Sample: X̄ = 5,50,000

              If Z is very large -> Unusual -> reject H0

              t-Test(Simple)

              When to use ?

                Sample size is small (n < 30)
                AND σ is unknown .
              
             t = (X̄ - μ) / (s / √n) , use s (sample std) instead of σ

             Meaning :-  
             Same idea:
                t = how far sample mean is from expected value

            But :     
                 More Uncertainty -> Wider distribution .


``Key different : - ``
              
              | Feature     | Z-test       | t-test            |
              | ----------- | ------------ | ----------------- |
              | Std dev     | σ known      | σ unknown (use s) |
              | Sample size | Large        | Small             |
              | Accuracy    | More precise | More conservative |


``Real problem and solved by Hypothesis all method .``
                
       A school says:

            “Our new teaching method increases average marks    

            above 60”

            You collect student marks:

            Data:-  [62, 64, 63, 65, 61, 66, 67]

            What is Hypothesis ?
                 
                 Hypothesis = a claim we want to test using data

              In simple words:
                     
                     We are not sure if something is true, so we test it.

                Create 2 Hypothesis :-    

                     Null-Hypothesis :-  Default assumption(nothing special)

                     H₀: Average marks = 60 (no improvement)
                     
                     
                     Alternative-Hypothesis :-  What we want to prove .

                     H₁: Average marks > 60 (improvement exists)

               IMPORTANT IDEA

                      We always assume H₀ is true first

                      Then we try to break it using data


                Look at the Data : -   
                      
                      your data :    
                         
                         [62, 64, 63, 65, 61, 66, 67]

                      average (mean) = 64

                      Now question is :

                      64 looks bigger than 60 but is it really significant .

                      or just random variation ?

                      This is where p-value comes :-
                      
                      What is p-value ?

                      “If H₀ is true (mean = 60), what is the probability of getting a result like this (mean = 64)?”

                      Two possibilities:

                      p-value is SMALL (e.g., 0.01)

                      THis result is VERY RARE if H0 is true , So H0 is probably wrong .

                      Reject H0
                      Accept H1 (improvement is real)

                      Case 2: p-value is Big (e.g., 0.30)

                      This result is common even if H0 is true .

                      So nothing special

                      Do not reject H0

                      Simple meaning (Never Forget)

                      p-value = "How surprising is my data ? "

                         Small → surprising → real effect
                         Large → normal → maybe luck

                      Where do t-test / z-test come ?

                      Now we need to calculate this p-value

                      We don't guess it .

                      t-test and z-test = tools .

                      test statistic → then → p-value

                      they help us compute:

                      test statistics -> then -> p-value

                      Which one do we use here ?

                      Small sample (7 values)
                      Population standard deviation unknown .

                      So we use t-test

            

        More accurate understanding
                      
                      Situation:

                      Someone says: “My average marks = 90”

                      Step 1: Define hypothesis

                          H₀: mean = 90
                          H₁: mean ≠ 90

                      Step 2: Collect marks

                      Example:
                              [85, 88, 92, 87, 90]

                                Mean ≈ 88.4

                                Step 3: Now question:

                      “Is 88.4 close enough to 90 (normal variation)?

                       OR

                       Is it too different (something is wrong)?”

            This is where p-value comes .
                       
                       Not simpel matching


                But how i know and how calculate p-value ?

                       No (In real life)

                       You almost never calculate it by hand .

                       I use Python (scipy)
                       Excel
                       Statistical software

                How i know p-value is big or small ?
                       
                       You compare with a threshold

                Significance Level (α)
                       
                       Most common value:

                       α = 0.05


              Rule (VERY IMPORTANT)

                           If p-value < 0.05  SMALL  Reject H₀
                          
                           If p-value ≥ 0.05  BIG Do NOT reject H₀


                  p-value = 0.003

                  0.003 < 0.05
                  Very small
                  Reject H₀ -  Real effect

               Case 2:

                  p-value = 0.20

                  0.20 > 0.05
                  Big
                  Keep H₀ → Maybe luck

            But Why 0.05 ?
                  
                  It means :      
                         
                         We accept only 5% chance of being wrong .

                         So:
                            
                            0.05 = 5% error tolerance

                            Very common standard


                      

```python
import numpy as np
from scipy import stats

# Data
data = [62, 64, 63, 65, 61, 66, 67]

# Old mean
old_mean = 60

# Perform t-test
t_stat, p_value = stats.ttest_1samp(data, old_mean)

print("Sample Mean:", np.mean(data))
print("T-statistic:", t_stat)
print("P-value:", p_value)

# Decision
if p_value < 0.05:
    print("Reject H₀ - Teaching method works")
else:
    print("Fail to reject H₀ - No strong evidence")

#Reject Null Hypothesis
#Accept Alternative
```
